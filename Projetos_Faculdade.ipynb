```python
# Lista para guardar as notas
notas = []

print("Sistema de Gestão de Notas")
print("Digite as notas das 4 provas aplicadas. Quando terminar, digite 'fim'.\n")

while True:
    entrada = input("Nota: ")
    if entrada.lower() == "fim":  # Para o loop se digitar 'fim'
        break
    try:
        nota = float(entrada)  # Converte para número
        if 0 <= nota <= 10:    # Verifica se está no intervalo permitido
            notas.append(nota)
        else:
            print("⚠️ Nota deve ser entre 0 e 10.")
    except:
        print("⚠️ Entrada inválida.")

# Calcula média (ou 0 se não houver notas)
media = round(sum(notas)/len(notas), 2) if notas else 0

# Mostra o relatório final
print("\nNotas:", notas)
print("Média:", media)
print("Situação:", "Aprovado" if media >= 7 else "Reprovado")
```

```python
# ============================================
# Passo 1: Definir a classe Livro
# ============================================
class Livro:
    def __init__(self, titulo, autor, genero, quantidade):
        self.titulo = titulo
        self.autor = autor
        self.genero = genero
        self.quantidade = quantidade

    def __str__(self):
        return (
            f"Título: {self.titulo} | "
            f"Autor: {self.autor} | "
            f"Gênero: {self.genero} | "
            f"Quantidade: {self.quantidade}"
        )


# ============================================
# Passo 2: Criar a lista de livros
# ============================================
biblioteca = []


# ============================================
# Passo 3: Implementar funções para gerenciar os livros
# ============================================

def cadastrar_livro(titulo, autor, genero, quantidade):
    """Cadastra um novo livro na biblioteca"""
    livro = Livro(titulo, autor, genero, quantidade)
    biblioteca.append(livro)


def listar_livros():
    """Lista todos os livros cadastrados"""
    if not biblioteca:
        print("Nenhum livro cadastrado.")
    else:
        print("\n--- Lista de Livros ---")
        for livro in biblioteca:
            print(livro)


def buscar_livro(titulo):
    """Busca um livro pelo título"""
    for livro in biblioteca:
        if livro.titulo.lower() == titulo.lower():
            print("\nLivro encontrado:")
            print(livro)
            return
    print(f"\nLivro '{titulo}' não encontrado.")


# ============================================
# Passo 4: Utilizar a biblioteca Matplotlib
# ============================================
import matplotlib.pyplot as plt

def gerar_grafico():
    """Gera um gráfico com a quantidade de livros por gênero"""
    if not biblioteca:
        print("Nenhum livro cadastrado para gerar gráfico.")
        return

    # Conta quantidade de livros por gênero
    generos = {}
    for livro in biblioteca:
        generos[livro.genero] = generos.get(livro.genero, 0) + livro.quantidade

    # Cria gráfico de barras
    plt.bar(generos.keys(), generos.values(), color="skyblue")
    plt.title("Quantidade de Livros por Gênero")
    plt.xlabel("Gênero")
    plt.ylabel("Quantidade")
    plt.show()


# ============================================
# Exemplo de uso (teste rápido)
# ============================================

cadastrar_livro("Dom Casmurro", "Machado de Assis", "Romance", 3)
cadastrar_livro("1984", "George Orwell", "Distopia", 2)
cadastrar_livro("A Revolução dos Bichos", "George Orwell", "Fábula", 5)

listar_livros()
buscar_livro("1984")
buscar_livro("Senhora")
gerar_grafico()
```
